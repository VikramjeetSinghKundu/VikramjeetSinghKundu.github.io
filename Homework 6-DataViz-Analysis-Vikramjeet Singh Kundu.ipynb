```python
import pandas as pd

# Loading the UFO dataset from the URL
url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
buildings_data = pd.read_csv(url)
```

```python
buildings_data.head()
```

# Creating visualizations with Altair

## 1st plot: Distribution of buildings by county

```python
import altair as alt

# Grouping by county and count buildings
county_counts = buildings_data.groupby("County")["Location Name"].count().reset_index(name="building_count")

# Plotting building distribution by county
plot1 = alt.Chart(county_counts).mark_bar().encode(
    x=alt.X("County:N", sort="-y", title="County"),
    y=alt.Y("building_count:Q", title="Number of Buildings"),
    color=alt.Color("building_count:Q", scale=alt.Scale(scheme="blues"), title="Building Count")
).properties(
    title="Distribution of Buildings by County"
)
plot1.save('plot1.html')
```

## 2nd plot: Interactive building area analysis by year constructed

```python
# Creating an interactive scatter plot
selection = alt.selection_interval(encodings=["x"])  # Selecting a time range along the x-axis

plot2 = alt.Chart(buildings_data).mark_circle().encode(
    x=alt.X("Year Constructed:O", title="Year Constructed"),
    y=alt.Y("Square Footage:Q", title="Square Footage"),
    tooltip=["Location Name", "Year Constructed", "Square Footage"]
).add_selection(
    selection
).properties(
    title="Building Area by Year Constructed"
).interactive()

plot2.save('plot2.html')
```
